```python
import seaborn as sns

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Basic EDA
print(titanic.info())
print(titanic.describe())

# Central Tendency for Age
age_mean = titanic['age'].mean()
age_median = titanic['age'].median()
age_mode = titanic['age'].mode()[0]

# Central Tendency for Fare
fare_mean = titanic['fare'].mean()
fare_median = titanic['fare'].median()
fare_mode = titanic['fare'].mode()[0]

print("\nCentral Tendency for Age:")
print(f"Mean: {age_mean:.2f}, Median: {age_median:.2f}, Mode: {age_mode}")

print("\nCentral Tendency for Fare:")
print(f"Mean: {fare_mean:.2f}, Median: {fare_median:.2f}, Mode: {fare_mode}")
```
